```python
import random

def main():
    print("\nðŸŽ® Welcome to the Ultimate High Low Game! ðŸŽ®")
    print("********************************************\n")

    try:
        total_rounds = int(input("How many rounds would you like to play? ðŸŽ¯ "))
    except ValueError:
        print("âš ï¸ Invalid input. Defaulting to 5 rounds.")
        total_rounds = 5

    your_score = 0

    for i in range(total_rounds):
        print(f"\nðŸ”µ Round {i + 1} of {total_rounds}")

        computer_number = random.randint(1, 100)
        your_number = random.randint(1, 100)

        print(f"ðŸ‘‰ Your number is: {your_number}")

        choice = input("Do you think your number is 'higher' or 'lower' than the computer's number? ").strip().lower()

        higher_and_correct = choice == "higher" and your_number > computer_number
        lower_and_correct = choice == "lower" and your_number < computer_number

        if higher_and_correct or lower_and_correct:
            print(f"âœ… Correct! The computer's number was {computer_number}.")
            funny_positive = random.choice([
                "ðŸ”¥ You're on fire!",
                "ðŸ˜Ž Pro level guess!",
                "ðŸŽ¯ Bullseye!",
                "ðŸ’ª Strong move!"
            ])
            print(funny_positive)
            your_score += 1
        else:
            print(f"âŒ Oops! The computer's number was {computer_number}.")
            funny_negative = random.choice([
                "ðŸ˜… Better luck next time!",
                "ðŸ˜‚ That was a close miss!",
                "ðŸ¤” Maybe trust your instincts more!",
                "ðŸ˜¬ Not quite right!"
            ])
            print(funny_negative)

        print(f"ðŸ† Current score: {your_score}")
        print("-" * 40)

    print("\nðŸŽ‰ Game Over!")
    print(f"Your final score: {your_score} out of {total_rounds}")

    # Grade system
    percentage = (your_score / total_rounds) * 100
    if percentage == 100:
        grade = "ðŸ… Legend!"
    elif percentage >= 80:
        grade = "ðŸ¥‡ Excellent!"
    elif percentage >= 60:
        grade = "ðŸ¥ˆ Good Job!"
    elif percentage >= 40:
        grade = "ðŸ¥‰ Not Bad!"
    else:
        grade = "ðŸ˜¢ Needs Practice."

    print(f"Your grade: {grade}")

if __name__ == "__main__":
    main()
```
